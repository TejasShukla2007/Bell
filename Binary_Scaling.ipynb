```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
img = cv2.imread('/content/1.jpg')
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
lower = np.array([100,55,40])
upper = np.array([900,900,900])
mask = cv2.inRange(img,lower,upper)
plt.imshow(mask,'gray')
```

# **Contour** 

```python
contour,_ = cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
contour
```

```python
print(img.shape)
print(mask.shape)
```

```python
drawncontour = cv2.drawContours(img,contour,-1,(255,255,0), 2)
plt.imshow(drawncontour)
```

```python
maxcontour = max(contour,key=cv2.contourArea)
maxcontour
```

```python
x,y,w,h = cv2.boundingRect(maxcontour)
cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(img)
```

```python
img_cropped = img[y:y+h,x:x+w]
plt.imshow(img_cropped)
```
