```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
img = cv2.imread('/content/handgesture.jfif')
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(img)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
cv2_imshow(img_gray)
```

```python
blur = cv2.blur(img_gray,(3,3))
```

```python
cv2_imshow(blur)
```

```python
r,thresh = cv2.threshold(blur,105,255,cv2.THRESH_BINARY)
```

```python
cv2_imshow(thresh)
```

```python
contours,_ = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img,contours,-1,(255,255,0),2)
```

```python
cv2_imshow(img)
```

```python
hull = [cv2.convexHull(conhull) for conhull in contours]
hull
```

```python
hulldrawn = cv2.drawContours(img,hull,-1,(0,0,255),2)
cv2_imshow(hulldrawn)
```

```python
hull_gray = cv2.drawContours(img_gray,hull,-1,(0,0,255),2)
cv2_imshow(hull_gray)
```

```python
hull_thresh = cv2.drawContours(thresh,hull,-1,(255,255,0),2)
cv2_imshow(hull_thresh)
```
