```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python
from tensorflow.keras.preprocessing import image
```

```python
img_1 = image.load_img("/content/1.jpg")
```

```python
plt.imshow(img_1)
```

```python
gray_img = cv2.imread('/content/1.jpg',0)
cv2_imshow(gray_img)
```

```python
plt.imshow(gray_img)
```

```python
r,thresh1 = cv2.threshold(gray_img, 120, 255, cv2.THRESH_BINARY)
cv2_imshow(thresh1)
```

```python
r,thresh2 = cv2.threshold(gray_img, 120,255,cv2.THRESH_BINARY_INV)
cv2_imshow(thresh2)
```

```python
r,thresh3 = cv2.threshold(gray_img, 120,255,cv2.THRESH_TRUNC)
cv2_imshow(thresh3)
```

```python
r,thresh4 =  cv2.threshold(gray_img, 120,255,cv2.THRESH_TOZERO)
cv2_imshow(thresh4)
```

```python
r,thresh5 =  cv2.threshold(gray_img, 120,255,cv2.THRESH_TOZERO_INV)
cv2_imshow(thresh5)
```

```python
r,thresh6 =  cv2.threshold(gray_img, 120,255,cv2.THRESH_OTSU)
cv2_imshow(thresh6) 
```
